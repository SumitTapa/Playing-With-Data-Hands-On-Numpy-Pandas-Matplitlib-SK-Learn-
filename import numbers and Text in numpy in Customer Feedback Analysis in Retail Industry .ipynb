```python
# Step 2: Extract Numerical Ratings for Analysis
ratings = data[:, 1].astype(int)  # Convert rating column to integers
print(f"Total Feedback Entries: {len(ratings)}")
```

```python
# Step 3: Perform Statistical Analysis
average_rating = np.mean(ratings)
print(f"Average Customer Rating: {average_rating:.2f}")
```

```python
highest_rating = np.max(ratings)
print(f"Highest Rating Received: {highest_rating}")

lowest_rating = np.min(ratings)
print(f"Lowest Rating Received: {lowest_rating}\n")
```

```python
unique_ratings, rating_counts = np.unique(ratings, return_counts=True)
print(f"Unique Ratings: {unique_ratings}, Unique Rating Counts: {rating_counts}")

# unique_ratings: An array of unique ratings (e.g., [1, 2, 3, 4, 5]).

# rating_counts: An array of how many times each rating appears (e.g., [1, 1, 1, 2, 3]).
    
```

```python
%pip install matplotlib
import matplotlib.pyplot as plt
```

```python
# Step 4: Extract Text Feedback for Qualitative Insights
feedback_comments = data[:, 2]

# Step 5: Display Insights
print("\n**Customer Feedback Analysis Summary**")
print()

print("**Rating Distribution:**")
print()
for rating, count in zip(unique_ratings, rating_counts):
    print(f"Rating {rating}: {count} customers")
```

# Visualizing the distributions of data using various plots

```python
# Create a bar plot for rating distribution
plt.figure(figsize=(10, 6))

# Bar plot
bars = plt.bar(unique_ratings, rating_counts, color='skyblue', edgecolor='navy', alpha=0.7)

# Customize the plot
plt.xlabel('Ratings', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)
plt.title('Distribution of Customer Ratings', fontsize=14, fontweight='bold')

# Add value labels on top of each bar
for bar, count in zip(bars, rating_counts):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(count)}', ha='center', va='bottom', fontsize=11)

# Add grid for better readability
plt.grid(axis='y', alpha=0.3, linestyle='--')

# Show the plot
plt.tight_layout()
plt.show()

```

```python
# Optional: Create a pie chart for percentage distribution
plt.figure(figsize=(8, 8))
colors = plt.cm.Set3(np.linspace(0, 1, len(unique_ratings)))

# Create pie chart
plt.pie(rating_counts, labels=[f'Rating {r}' for r in unique_ratings], 
        autopct='%1.1f%%', colors=colors, startangle=90)
plt.title('Customer Ratings Percentage Distribution', fontsize=14, fontweight='bold')

# Add a circle at the center to create a donut chart (optional)
centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Additional statistics
print("\n**Additional Statistics:**")
print(f"Total number of reviews: {sum(rating_counts)}")
print(f"Average rating: {np.average(unique_ratings, weights=rating_counts):.2f}")
"""max() returns the maximum value itself
argmax() returns the index (position) where the maximum value occurs"""
print(f"Most common rating: {unique_ratings[np.argmax(rating_counts)]}")

# Create a histogram of feedback comment lengths
comment_lengths = [len(str(comment).split()) for comment in feedback_comments]

plt.figure(figsize=(10, 6))
plt.hist(comment_lengths, bins=20, color='lightcoral', edgecolor='darkred', alpha=0.7)
plt.xlabel('Number of Words in Feedback', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Feedback Comment Lengths', fontsize=14, fontweight='bold')
plt.grid(axis='y', alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()

print(f"\nAverage comment length: {np.mean(comment_lengths):.1f} words")
print(f"Shortest comment: {min(comment_lengths)} words")
print(f"Longest comment: {max(comment_lengths)} words")
```
